```python
covid_df.drop([ "Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True, axis = 1  )
```

```python
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime (covid_df['Date'],format = '%Y-%m-%d')
```

```python
#active cases
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc="max")
```

```python
statewise["Recovery Rate"] = statewise ["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise ["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient (cmap = "GnBu")
```

```python
Top_10_Active_Cases = covid_df.groupby('State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by=['Active_Cases'], ascending=False).reset_index()
```

```python
fig = plt.figure(figsize = (16,9))
```

```python
plt.title ("Top 10 state most active  cases in India",size = 25)
```

```python
ax = sns.barplot( data = Top_10_Active_Cases.iloc[:10], y = "Active_Cases",x ="State/UnionTerritory", linewidth = 2, edgecolor = 'red', hue='State/UnionTerritory' )
```

```python
Top_10_Active_Cases = covid_df.groupby('State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by=['Active_Cases'], ascending=False).reset_index()

fig = plt.figure(figsize = (16,9))
plt.title ("Top 10 state most active  cases in India",size = 25)

ax = sns.barplot( data = Top_10_Active_Cases.iloc[:10], y = "Active_Cases",x ="State/UnionTerritory", linewidth = 2, edgecolor = 'red', hue='State/UnionTerritory' )
Top_10_Active_Cases.style.background_gradient (cmap = "GnBu")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# Top states with highest deaths
top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by=['Deaths'], ascending=False).reset_index()

# Plotting
fig = plt.figure(figsize=(19, 6))
plt.title("Top 10 states with the most deaths in India", size=25)

ax = sns.barplot(data=top_10_deaths.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth=2, edgecolor='red', hue='State/UnionTerritory')
top_10_deaths.style.background_gradient(cmap="GnBu")
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
fig = plt.figure(figsize=(19, 6))
ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Uttar Pradesh', 'Tamil Nadu'])], x='Date', y='Active_Cases', hue='State/UnionTerritory')
ax.set_title("Top 5 Affected States in India", size=16)
plt.show()
```

```python
#active cases
country_wise_df['Active_Cases'] = country_wise_df['Confirmed'] - (country_wise_df['Recovered'] + country_wise_df['Deaths'])
country_wise_df.tail()
```

```python
country_wise = pd.pivot_table(country_wise_df, values=["Confirmed", "Deaths", "Recovered"], index="Country/Region", aggfunc="max")
```

```python
fig = plt.figure(figsize = (16,9))
```

```python
plt.title ("Top 10 country most active  cases in world",size = 25)
```

```python
country_wise["Recovery Rate"] = country_wise ["Recovered"]*100/country_wise["Confirmed"]
```

```python
country_wise["Mortality Rate"] = country_wise ["Deaths"]*100/country_wise["Confirmed"]
```

```python
country_wise = country_wise.sort_values(by = "Confirmed", ascending = False)
```

```python
country_wise.style.background_gradient (cmap = "GnBu")
```

```python
Top_10_country_Cases = country_wise_df.groupby('Country/Region').max()[['Active_Cases', 'Active']].sort_values(by=['Active_Cases'], ascending=False).reset_index()
```

```python
ax = sns.barplot( data = Top_10_country_Cases.iloc[:10], y = "Active_Cases",x ="Country/Region", linewidth = 2, edgecolor = 'red', hue='Country/Region' )
```

```python
Top_10_country_Cases = country_wise_df.groupby('Country/Region').max()[['Active_Cases', 'Active']].sort_values(by=['Active_Cases'], ascending=False).reset_index()

fig = plt.figure(figsize = (16,9))
plt.title ("Top 10 country most active  cases in the world",size = 25)

ax = sns.barplot( data = Top_10_country_Cases.iloc[:10], y = "Active_Cases",x ="Country/Region", linewidth = 2, edgecolor = 'red', hue='Country/Region' )
Top_10_country_Cases.style.background_gradient (cmap = "GnBu")
plt.xlabel("Countries")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
Top_10_country_deaths_Cases = country_wise_df.groupby('Country/Region').max()[['Deaths', 'New deaths']].sort_values(by=['Deaths'], ascending=False).reset_index()

fig = plt.figure(figsize = (16,9))
plt.title ("Top 10 country most active deaths  cases in the world",size = 25)

ax = sns.barplot( data = Top_10_country_Cases.iloc[:10], y = "Deaths",x ="Country/Region", linewidth = 2, edgecolor = 'red', hue='Country/Region' )
Top_10_country_deaths_Cases.style.background_gradient (cmap = "GnBu")
plt.xlabel("Countries")
plt.ylabel("Total deaths Cases")
plt.show()
```
